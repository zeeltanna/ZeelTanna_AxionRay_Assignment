```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel("D:\Axion Ray\DA -Task 2..xlsx")
df
```

```python
df.isnull().sum()
```

```python
df.drop(columns="CAMPAIGN_NBR", inplace=True)
df
```

```python
columns_to_fill = ["ENGINE_SOURCE_PLANT", "ENGINE_TRACE_NBR", "TRANSMISSION_SOURCE_PLANT", "TRANSMISSION_TRACE_NBR"]
df[columns_to_fill] = df[columns_to_fill].fillna("Unknown")
df.isnull().sum()
```

```python
columns = ["OPTN_FAMLY_CERTIFICATION", "OPTF_FAMLY_EMISSIOF_SYSTEM"]
df[columns] = df[columns].fillna("Undefined")
df
```

```python
columns_to_fill = ["CAUSAL_PART_NM", "PLANT","VEH_TEST_GRP"]
df[columns_to_fill] = df[columns_to_fill].fillna("other")
df.isnull().sum()
```

```python
columns_to_fill = ["STATE", "REPAIR_DLR_POSTAL_CD"]
df[columns_to_fill] = df[columns_to_fill].fillna("unknown")
df.isnull().sum().sum()
```

```python
columns_to_fill = ["LINE_SERIES", "LAST_KNOWN_DELVRY_TYPE_CD"]
df[columns_to_fill] = df[columns_to_fill].fillna("unknown")
df.isnull().sum().sum()
```

```python
df["TOTALCOST"] = pd.to_numeric(df["TOTALCOST"], errors='coerce')
df["TOTALCOST"] = df["TOTALCOST"].astype(float)
```

```python
df["REPORTING_COST"] = pd.to_numeric(df["REPORTING_COST"], errors='coerce')
df["REPORTING_COST"] = df["REPORTING_COST"].astype(float)
```

```python
df["TOTALCOST"] = df["TOTALCOST"].fillna(df["REPORTING_COST"])
df.isnull().sum().sum()
```

```python
df["LBRCOST"] = pd.to_numeric(df["LBRCOST"], errors='coerce')
df["LBRCOST"] = df["LBRCOST"].astype(float)
```

```python
df["KM"] = pd.to_numeric(df["KM"], errors='coerce')
df["KM"] = df["KM"].astype(int)
```

```python
df["REPAIR_DATE"] = pd.to_datetime(df["REPAIR_DATE"], errors='coerce')
```

```python
df["REPAIR_AGE"] = pd.to_numeric( df["REPAIR_AGE"], errors='coerce')
df["REPAIR_AGE"] = df["REPAIR_AGE"].astype(int)
```

```python
def generate_tags(row):
    tags = []
    if row['TOTALCOST'] > 10000:
        tags.append('HighRepairCost')
    if row['REPAIR_AGE'] > 500:
        tags.append("HighVehicleAge")
    if row['CAUSAL_PART_NM'] == 'ENGINE':
      tags.append('EngineFailure')
    if row['PLATFORM'] == 'A':
        tags.append('PlatformA')
    tags.append(row['BODY_STYLE'])

    return ','.join(tags)

df['Tags'] = df.apply(generate_tags, axis=1)

tag_counts = df['Tags'].str.get_dummies(sep=',').sum().sort_values(ascending=False)
print("Tag Counts:\n", tag_counts)

high_cost_repairs = df[df['Tags'].str.contains('HighRepairCost')]

print("\nBody Style vs. Failure Counts:")
body_style_failure = df.groupby('BODY_STYLE')['Tags'].apply(lambda x: x.str.split(',').explode().value_counts())
body_style_failure
```
